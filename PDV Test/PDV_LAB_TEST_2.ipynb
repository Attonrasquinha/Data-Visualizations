# Milk Production

```python
import pandas as pd
import numpy as np
df=pd.read_csv('/home/msois/Downloads/PDV/PDV Test/milk_production.csv')
df=df.applymap(lambda x: x.strip() if isinstance(x,str) else x)
df.head()
```

```python
df=df.fillna(0)
```

```python
df
```

# 1

```python
#Which state has produced max milk in year 2013-14?

df['total']=df['Cow Milk-2013-14']+df['Boffalo Milk-2013-14']+df['Goat Milk-2013-14']
```

```python
df['total']
```

```python
df[df["total"]==df["total"].max()]["State/ UT Name"]
```

# 2

```python
#List top 5 milk producing states in each year from 2010 to 2015
df["y2011"] =(df["Cow Milk-2010-11"] + df["Boffalo Milk-2010-11"] + df["Goat Milk-2010-11"])
df["y2012"]= (df["Cow Milk-2011-12"] + df["Boffalo Milk-2011-12"] + df["Goat Milk-2011-12"])
df["y2014"]= (df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"])
df["y2015"]= (df["Cow Milk-2014-15"] + df["Boffalo Milk-2014-15"] + df["Goat Milk-2014-15"])
df["y2016"]= (df["Cow Milk-2015-16"] + df["Boffalo Milk-2015-16"] + df["Goat Milk-2015-16"])
```

```python
df.sort_values('y2011',ascending=False)["State/ UT Name"].head()
```

```python
df.sort_values('y2012',ascending=False)["State/ UT Name"].head()
```

```python
df.sort_values('y2014',ascending=False)["State/ UT Name"].head()
```

```python
df.sort_values('y2015',ascending=False)["State/ UT Name"].head()
```

# 3

```python
#Calculate average milk production by cow, buffalo and goat in each year for selected state.
df["cow_all_year"]=df["Cow Milk-2010-11"]+df["Cow Milk-2011-12"]+df["Cow Milk-2013-14"]+df["Cow Milk-2014-15"]+df["Cow Milk-2015-16"]
df["Boffalo_all_year"]=df["Boffalo Milk-2010-11"]+df["Boffalo Milk-2011-12"]+df["Boffalo Milk-2013-14"]+df["Boffalo Milk-2014-15"]+df["Boffalo Milk-2015-16"]
df["Goat_all_year"]=df["Goat Milk-2010-11"]+df["Goat Milk-2011-12"]+df["Goat Milk-2013-14"]+df["Goat Milk-2014-15"]+df["Goat Milk-2015-16"]
```

```python
df[df["cow_all_year"].mean() and df["State/ UT Name"]=="Uttar Pradesh"]
```

```python
df[df["Boffalo_all_year"].mean() and df["State/ UT Name"]=="Uttar Pradesh"]
```

```python
df[df["Goat_all_year"].mean() and df["State/ UT Name"]=="Rajasthan"]
```

# 4

```python
#List 5 states (if present) whose total milk production is increased for last three years.
df[df["y2016"].ge(df["y2015"]) & df["y2015"].ge(df["y2014"])].head()
```

```python
#Draw line graph showing total milk production for last 5 years of selected state
import matplotlib.pyplot as plt
df['total_production']=df["cow_all_year"]+df["Boffalo_all_year"]+df["Goat_all_year"]
x=['Andhra Pradesh','Chhattisgarh','Assam','Gujarat']
plt.plot(x,df['total_production'].head(4), linestyle="dotted", marker='^', color="red", label = 'Product 1')
plt.xlabel("States")
plt.ylabel("Milk in Litres")



#title
plt.title("total milk production ")
plt.show()
```

```python
fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

x=['Andhra Pradesh','Chhattisgarh','Assam','Gujarat']


ax1.pie(df['cow_all_year'].head(4),labels=x)

ax2.pie(df['Boffalo_all_year'].head(4),labels=x)

ax3.pie(df['Goat_all_year'].head(4),labels=x)

#ax2.pie(x,y)
#ax3.pie()
plt.tight_layout()

plt.show()
```
