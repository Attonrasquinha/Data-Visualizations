```python
import pandas as pd
import numpy as np
df1=pd.read_csv('agri_data.csv')
df1=df1.dropna(how="all", axis=1)
df1=df1.fillna(0)
df1.head(3)
```

```python
df2=pd.read_csv('forest_data.csv')
df2=df2.dropna(how="all", axis=1)
df2=df2.fillna(0)
df2.head(3)
```

```python
df3=pd.read_csv("Metadata_country_agri.csv")
df4=pd.read_csv("Metadata_country_forest.csv")
df4.head(3)
```

```python
agri_merge=pd.merge(df3,df1,on="Country Code",how="left")
agri_merge.head(3)
```

```python
forest_merge=pd.merge(df4,df2,on="Country Code",how="left")
forest_merge.head()
```

```python
agri_merge.columns
```

```python
forest_merge.columns
```

```python
agri_merge["Region"].unique()
```

```python
forest_merge["Region"].unique()
```

```python
agri_merge["totalarea"]=agri_merge.iloc[:,7:].mean(axis=1)
```

```python
agri_merge.head()
```

```python
agri_merge.sort_values("totalarea",ascending=False).groupby("Region")["Country Name"].first()
```

```python
forest_merge["totalarea"]=forest_merge.iloc[:,8:].mean(axis=1)
forest_merge["dec"]=forest_merge.loc[:,"1990"]-forest_merge.loc[:,"2016"]
a=forest_merge.sort_values("dec").groupby("Region",as_index=False)[["Region","Country Name"]].head()
a=a.dropna(0)
a.sort_values("Region")
```

```python
xl=agri_merge.groupby("Region",as_index=False)["totalarea"].mean()["Region"]
xl
```

```python
y1=forest_merge.groupby("Region",as_index=False)["totalarea"].mean()["Region"].values
y1
```

```python
x=agri_merge.groupby("Region")["totalarea"].mean().values
x
```

```python
y=forest_merge.groupby("Region")["totalarea"].mean().values
y
```

```python
label=np.array(["agriculture","forest"])
```

```python
import matplotlib.pyplot as plt 
import numpy as np
fig = plt.figure(figsize=(8,8))
ax1=fig.add_subplot(4,2,1)
ax2=fig.add_subplot(4,2,2)
ax3=fig.add_subplot(4,2,3)
ax4=fig.add_subplot(4,2,4)
ax5=fig.add_subplot(4,2,5)
ax6=fig.add_subplot(4,2,6)
ax7=fig.add_subplot(4,2,7)

ax1.pie(np.array([x[0],y[0]]),labels=label,autopct="%1.1f%%")
ax1.set_title(xl[0])
ax1.axis("equal")

ax2.pie(np.array([x[1],y[1]]),labels=label,autopct="%1.1f%%")
ax2.set_title(xl[1])
ax2.axis("equal")

ax3.pie(np.array([x[2],y[2]]),labels=label,autopct="%1.1f%%")
ax3.set_title(xl[2])
ax3.axis("equal")

ax4.pie(np.array([x[3],y[3]]),labels=label,autopct="%1.1f%%")
ax4.set_title(xl[3])
ax4.axis("equal")

ax5.pie(np.array([x[4],y[4]]),labels=label,autopct="%1.1f%%")
ax5.set_title(xl[4])
ax5.axis("equal")

ax6.pie(np.array([x[5],y[5]]),labels=label,autopct="%1.1f%%")
ax6.set_title(xl[5])
ax6.axis("equal")

ax7.pie(np.array([x[6],y[6]]),labels=label,autopct="%1.1f%%")
ax7.set_title(xl[6])
ax7.axis("equal")

plt.tight_layout()
plt.show()
```

```python
forest_merge["first5year"]=forest_merge.loc[:,"1990":"1995"].mean(axis=1)
forest_merge["last5year"]=forest_merge.loc[:,"2012":"2016"].mean(axis=1)
```

```python
x=forest_merge.groupby("Region")["first5year"].mean().values
x
```

```python
y=forest_merge.groupby("Region")["last5year"].mean().values
y
```

```python
xlabel=forest_merge.groupby("Region",as_index=False)["first5year"].mean()["Region"].values
xlabel
```

```python
plt.bar(np.arange(1,8),x,width=0.2,label="Agriculture")
plt.bar(np.arange(1,8)+0.25,y,width=0.25,label="Forest")
plt.xlabel("Region")
plt.ylabel("average")
plt.legend()
plt.xticks(np.arange(1,8),(list(xlabel)),rotation=60)
plt.show()
```
